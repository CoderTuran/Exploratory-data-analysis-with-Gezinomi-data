```python
# verisetine genel bir bakış
def check_df(dataframe, head=5):
    print("##################### Shape #####################")
    print(dataframe.shape)
    print("##################### Types #####################")
    print(dataframe.dtypes)
    print("##################### NA #####################")
    print(dataframe.isnull().sum())
    print("##################### Quantiles #####################")
    print(dataframe.describe([0, 0.05, 0.50, 0.95, 0.99, 1]).T)
```

```python
# tüm kolonlar için unique değerler ve value_counts kontrolü :
def summary_value_counts_with_unique_count(df):
    for column in df.columns:
        unique_count = df[column].nunique()
        display(column)
        display(f"Eşsiz değer sayısı: {unique_count}")
        display(df[column].value_counts())
        display("-------------------------------------------------")
```

```python
# object olarak görünen tarih içeren kolonlar date olarak değiştirildi
df["SaleDate"] = pd.to_datetime(df["SaleDate"]).dt.normalize()
df["CheckInDate"] = pd.to_datetime(df["CheckInDate"]).dt.normalize()
df["CheckOutDate"] = pd.to_datetime(df["CheckOutDate"]).dt.normalize()
```

```python
check_df(df)
```

```python
# yukarıda, check in date içinde min olarak görülebilen 2007-07-20 00:00:00 tarihi
# satış-checkout tarihine göre 2020-07-20 00:00:00 olarak revize edildi
df["CheckInDate"].min()
```

```python
df.loc[df["CheckInDate"] == "2007-07-20 00:00:00", "CheckInDate"] = "2020-07-20 00:00:00"
df.loc[df["CInYear"] == "2007", "CInYear"] = 2020
```

```python
# satis_checkin_day_diff kolonunda 0'dan küçük değerler olduğu görüldü. satış tarihi check-in tarihinden sonra olamayacağı için
# bu satırlar için satış tarihi check in tarihine eşitlenmesine karar verildi.
def update_for_sale_checkin_diff(df):
    rows_to_update = df[df["satis_checkin_day_diff"] < 0].index
    df.loc[rows_to_update, "CheckInDate"] = df.loc[rows_to_update, "SaleDate"]
    df["satis_checkin_day_diff"] = (pd.to_datetime(df["CheckInDate"]) - pd.to_datetime(df["SaleDate"])).dt.days
    return df
```

```python
# ödeme şekli havale ve nakit olan satırlarda, InstallmentNumber-ShowInstallmentNumber-CardName kolonları için veriler
# Null geliyor, bunlar peşin ödeme olduğu için bu değerler 0 ile güncellenmesine karar verildi.
def update_havale_to_zero(df):
    condition = df["PayType"] == "Havale"
    columns_to_update = ["InstallmentNumber", "ShowInstallmentNumber", "CardName"]
    df.loc[condition, columns_to_update] = 0
    return df

def update_nakit_to_zero(df):
    condition = df["PayType"] == "Nakit"
    columns_to_update = ["InstallmentNumber", "ShowInstallmentNumber", "CardName"]
    df.loc[condition, columns_to_update] = 0
    return df
```
